# Model Building and prediction
The algorithms that are used here to try to do the anomaly detection on this dataset are

Isolation Forest Algorithm
Local Outlier Factor(LOF) Algorithm and also
Support Vector Machine (SVM) Algorithm

```python
#Define the outlier detection method

classifiers = {
    "Isolation Forest":IsolationForest(n_estimators=100, max_samples=len(X), 
                                       contamination=outlier_fraction,random_state=state, verbose=0),
    "Local Outlier Factor":LocalOutlierFactor(n_neighbors=20, algorithm='auto', 
                                              leaf_size=30, metric='minkowski',
                                              p=2, metric_params=None, contamination=outlier_fraction),
    "Support Vector Machine":OneClassSVM(kernel='rbf', degree=3, gamma=0.1,nu=0.05, 
                                         max_iter=-1)
   
}
```

```python
n_outliers = len(Fraud)
for i ,(clf_name ,clf) in enumerate(classifiers.items()) :

  if clf_name == 'Isolation Forest':
    clf.fit(X)
    y_pred = clf.predict(X)
    scores_prediction = clf.decision_function(X)

  elif clf_name == 'Local Outlier Factor' :
    y_pred = clf.fit_predict(X)
    scores_prediction = clf.negative_outlier_factor_
  else:
     clf.fit(X)
     y_pred = clf.predict(X)

    #Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions

  y_pred[y_pred == 1] = 0
  y_pred[y_pred == -1] = 1
  n_errors = (y_pred != Y).sum()

 #Run the classification metrics

  print("{}: {}".format(clf_name,n_errors))
  print("Accuracy Score :")
  print(accuracy_score(Y,y_pred))
  print("Classification Report :")
  print(classification_report(Y,y_pred))
```

## Conclusion And Observation

- **Isolation Forest** detected 73 errors, outperforming **Local Outlier Factor (LOF)** with 97 errors and **Support Vector Machine (SVM)** with 8516 errors.

- The **Isolation Forest** demonstrated a significantly higher accuracy rate of **99.74%**, surpassing LOF (99.65%) and SVM (70.09%).

- Examining precision and recall for error detection:

  - **LOF:** 2% detection rate.

- Overall, the **Isolation Forest** method proved notably effective in identifying fraud cases, achieving an approximate **30%** detection rate.
